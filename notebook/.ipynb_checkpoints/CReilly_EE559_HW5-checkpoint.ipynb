```python
# Part C - Feature Extraction (using trainingData and testData)

# Part Ci: List time-domain features used in time-series classification. They include:
#     mean, minimum, maximum, std dev, skewness, kurtosis

# Part Cii: Extract minimum, maximum, mean, median, std dev, 1st quartile, 3rd quartile, for all 6 time series

instanceMx = np.dstack((trainingData,testData)); # create mx for for loop - instance determination

n = 88; # number of instances
z = 6; # number of time series
l = 7; # number of time series features
featureExtraction = np.zeros([n,l*z+1]);

for i in np.arange(n): # for-loop for n rows
    featureExtraction[i,0] = i+1; # define instance number
    
    for j in np.arange(z): # for loop for each time series
        featureExtraction[i,l*j+1] = instanceMx[:,j+1,i].min(); # min of time series
        featureExtraction[i,l*j+2] = instanceMx[:,j+1,i].max(); # max of time series
        featureExtraction[i,l*j+3] = instanceMx[:,j+1,i].mean(); # mean of time series
        featureExtraction[i,l*j+4] = np.median(instanceMx[:,j+1,i]); # median of time series
        featureExtraction[i,l*j+5] = instanceMx[:,j+1,i].std(); # std of time series
        featureExtraction[i,l*j+6] = np.quantile(instanceMx[:,j+1,i], 0.25); # 1st quartile of time series
        featureExtraction[i,l*j+7] = np.quantile(instanceMx[:,j+1,i], 0.75); # 3rd quartile of time series

# Create the column headers in for-loop and create dataframe
cols = 'Instance';
for i in np.arange(z):
    cols = np.hstack((cols,'min' + str(i+1),'max' + str(i+1),'mean' + str(i+1),'med' + str(i+1),'std' + str(i+1),
                     '1stqrt' + str(i+1),'3rdqrt' + str(i+1),));

instanceDF = pd.DataFrame(data=featureExtraction,columns=cols);

# Print resultant dataframe
instanceDF
```

```python
# Part Ciii - estimate std deviation of each column
k = featureExtraction.shape[1]-1;
n = featureExtraction.shape[0];
standardDeviation = np.zeros(k);

for i in np.arange(k):
    feature = featureExtraction[:,i+1];
    standardDeviation[i] = sum((feature - feature.mean())**2)/(n-8-1);
print('Estimated Std Dev for each instance = \n\n',standardDeviation)
```

```python
# Part Ciii - Bootstrap to see confidence interval of std. deviation

# Want to perform 20 bootstrap samples, so we can then choose the 5% delta from mean and the 95% delta from the mean,
# which will give us a 90% interval.

# draw 88 samples for each bootstrap set to determine std deviation...
z = 20; # how many bootstrap datasets we need...
q = featureExtraction.shape[1]-1; # number of features we are bootstrapping
n = featureExtraction.shape[0]; # number of samples for each bootstrap set
l = np.arange(n); # possible instance values

bootstrap = np.zeros(featureExtraction.shape) # create memory loc for bootstrap featureExtraction
bootstrapDev = np.zeros([q,z]) # create memory for Std Dev of each sample
deltas = bootstrapDev; # same set of zeros to start out
for i in np.arange(z): # perform 20 bootstrap samples
    
    v = rand.choices(l, k=n); # create array of new sample indices, with replacement
    for j in np.arange(n):
        bootstrap[j,:] = featureExtraction[v[j],:]; # create bootstrap set
    
    # Capture standard deviation of bootstrap
    for u in np.arange(q):
        feature = bootstrap[:,u+1];
        bootstrapDev[u,i] = sum((feature - feature.mean())**2)/(n-8-1);

for i in np.arange(z): # for loop to calculate deltas
    deltas[:,i] = standardDeviation - bootstrapDev[:,i];

deltas = np.sort(deltas,axis=1) # sort each column from lowest to highest.

output = np.zeros([q,2])
output[:,0] = standardDeviation.T-deltas[:,18]
output[:,1] = standardDeviation.T-deltas[:,1]

print('---Confidence Interval for Standard Deviation---\n')
print('90% Bootstrap Confidence Int. [5%, 95%] :\n')

print('\n', output)
```

```python
# Part 8aii

print('Part 8aii : Optimal (l,p) pairs:\n')

f = 20; # number of different models we want
y_train = np.zeros([f,len(trainingLabel)])
y_test = np.zeros([f,len(testLabel)])
p_none = np.zeros(f)
# logRegCoeffs = np.zeros(f,)

for a in np.arange(f): # Create for-loop for 20 time series
    g = a+1; # running out of variables!! How many 'new sets' the old set is split into...
    newLength = int(np.floor(trainingData.shape[0]/g));

    trainingData2 = trainingData[:newLength,:,:];
    testData2 = testData[:newLength,:,:];
    
    if g > 1:
        for i in np.arange(a):
            trainingData2 = np.hstack((trainingData2,trainingData[(i+1)*newLength:(i+2)*newLength,1:,:]))
            testData2 = np.hstack((testData2,testData[(i+1)*newLength:(i+2)*newLength,1:,:]))

    # With newly manipulated time series, find new features...
    
    instanceMx = np.dstack((trainingData2,testData2)); # create mx for for loop - instance determination

    n = 88; # number of instances in training data
    z = 6*g; # number of time series
    l = 7; # number of time series features
    featureExtraction = np.zeros([n,l*z+1]);

    for i in np.arange(n): # for-loop for n rows
        featureExtraction[i,0] = i+1; # define instance number

        for j in np.arange(z): # for loop for each time series
            featureExtraction[i,l*j+1] = instanceMx[:,j+1,i].min(); # min of time series
            featureExtraction[i,l*j+2] = instanceMx[:,j+1,i].max(); # max of time series
            featureExtraction[i,l*j+3] = instanceMx[:,j+1,i].mean(); # mean of time series
            featureExtraction[i,l*j+4] = np.median(instanceMx[:,j+1,i]); # median of time series
            featureExtraction[i,l*j+5] = instanceMx[:,j+1,i].std(); # std of time series
            featureExtraction[i,l*j+6] = np.quantile(instanceMx[:,j+1,i], 0.25); # 1st quartile of time series
            featureExtraction[i,l*j+7] = np.quantile(instanceMx[:,j+1,i], 0.75); # 3rd quartile of time series

    # Process data to pull out bending motion to get color difference...

    trainingFeatures = featureExtraction[:69,1:]; # grab time-domain features of training data
    testFeatures = featureExtraction[69:,1:]; # grab time-domain features of test data
    
    # Solve the optimal p value using Stratified 5-fold cross validation...
    
    svc = SVC(kernel='linear');
    rfecv = RFECV(estimator=svc, step=5, cv=StratifiedKFold(n_splits=5), scoring='accuracy', min_features_to_select = 1);
    rfecv.fit(trainingFeatures,trainingLabel_bin);
    trainingFeatures = rfecv.transform(trainingFeatures);
    testFeatures = rfecv.transform(testFeatures);
    
    if g == 1:
        train8vi = trainingFeatures;
    
    p_none[a] = rfecv.n_features_; # capture the optimal number of features

    print('(',str(g),', ',str(p_none[a]), ')'); # print optimal (l,p) pairing
    
    logReg = LogisticRegression(penalty = 'none', max_iter = 100); # create Logistic Regression class
    logReg.fit(trainingFeatures,trainingLabel_bin);

    
    # Print Coefficients
    for i in np.arange(logReg.coef_.shape[1]):
        print('\tB', str(i+1), ' = ', logReg.coef_[0,i])
    
    y_train[a,:] = logReg.predict(trainingFeatures);
    y_test[a,:] = logReg.predict(testFeatures);   
    
```

```python
# Part 8aiii - Print Confusion Matrices
# Part 8aiv - Compare MSE Error of Training and Test Sets

print('Part 8aiii : Print Confusion Matrices')
print('Part 8aiv : Accuracy Comparison: [Training MSE, Test MSE]\n')

totalError8a = 0;

for l in np.arange(f):
    cf = met.confusion_matrix(y_train[l,:],trainingLabel_bin)
    print('Confusion Matrix, L = ', str(l+1), ' : \n', cf, '\n')
    
    print('Error Data :\t [', str(sum(abs(y_train[l,:] - trainingLabel_bin))), ' , ', 
          str(sum(abs(y_test[l,:] - testLabel_bin))), ' ]\n')
    
    totalError8a = totalError8a + sum(abs(y_test[l,:] - testLabel_bin));
    
print('Total Error = ', totalError8a)
       
```

```python
# Part 8cii - Naive Bayes Classifier, Gaussian pdf

def NaiveBayesGaussian(X_train,y_train,X_test,y_test,q): 
    # Description
    # y_train, y_test must be integers
    # X_train, X_test must be same length as corresponding y's
    # Need to filter out redundant features prior
    import numpy as np
    import pandas as pd
    
    # Step 1 : create mxs for each class to get means, std.
    #q = 6; # number of classes
    
    # Class 0 through n - training
    len_y = np.arange(y_train.shape[0]);
    for j in np.arange(q):
        expression = 'c' + str(j) + '_train = np.zeros([1,X_train.shape[1]])';
        exec(expression);
        
        expression = 'c' + str(j) + '_train = np.vstack([c' + str(j) + '_train,X_train[i,:]])'
        for i in len_y:
            if y_train[i] == j:
                exec(expression);
                
        expression = 'c' + str(j) + '_train = np.delete(c' + str(j) + '_train,0,0)'
        exec(expression);
    
    # Class 0 through n - test
    len_y = np.arange(y_test.shape[0]);
    for j in np.arange(q):
        expression = 'c' + str(j) + '_test = np.zeros([1,X_test.shape[1]])';
        exec(expression);
        
        expression = 'c' + str(j) + '_test = np.vstack([c' + str(j) + '_test,X_test[i,:]])'
        for i in len_y:
            if y_train[i] == j:
                exec(expression);
                
        expression = 'c' + str(j) + '_test = np.delete(c' + str(j) + '_test,0,0)'
        exec(expression);
  
    # Step 2 : Create a priori probabilities
    aprob = np.zeros(q);
    for j in np.arange(q):
        expression = 'aprob[j] = c' + str(j) + '_train.shape[0]/y_train.shape[0]';
        exec(expression);
    
    # Step 3 : Calculate mean and standard deviation for each conditional class - training set only!
    
    p_train = np.zeros([X_train.shape[1],2,q]); # [mean, std]
    counter = np.arange(X_train.shape[1])
    
    for j in np.arange(q): 
        for i in counter:
            expression = 'p_train[i,0,' + str(j) + '] = c' + str(j) + '_train[:,i].mean()'
            exec(expression)
            expression = 'p_train[i,1,' + str(j) + '] = c' + str(j) + '_train[:,i].std()*np.sqrt(c' + str(j) + '_train.shape[0]/(c' + str(j) + '_train.shape[0]-1))'
            exec(expression)
        
    # Step 4 : Create probability Matrices from Training Set - assuming Conditional Gaussian
    
    prob_train = np.zeros([X_train.shape[0],X_train.shape[1],q]);
    #num = np.zeros([X_train.shape[0],q]);
    counter = np.arange(X_train.shape[1]) # create counter for the for loop
    for i in counter: # for loop i for each feature
        
        # Calculate Probability of Training Mx
        for j in np.arange(q):
            num = np.array(-0.5*((X_train[:,i] - p_train[i,0,j])/p_train[i,1,j])**2);
            prob_train[:,i,j] = 1/(np.sqrt(2*np.pi*p_train[i,1,j]**2))*np.exp(num);
    
    prob_test = np.zeros([X_test.shape[0],X_test.shape[1],q]);
    #num = np.zeros([X_test.shape[0],q]);                
    counter = np.arange(X_test.shape[1]) # create counter for the for loop
    for i in counter: # for loop i for each feature
        
        # Calculate Probability of Test Mx
        for j in np.arange(q):
            num = np.array(-0.5*((X_test[:,i] - p_train[i,0,j])/p_train[i,1,j])**2);
            prob_test[:,i,j] = 1/(np.sqrt(2*np.pi*p_train[i,1,j]**2))*np.exp(num);  
        
    # Step 5 : Classify predicts based on Naive Bayes classification
    
    # Training Set
    y_train_predict = np.zeros(y_train.shape[0]);
    counter = np.arange(y_train.shape[0]);
    condProb = np.zeros(q);
    trainErrorMSE = 0;
    for i in counter:
        for j in np.arange(q):
            condProb[j] = np.prod(prob_train[i,:,j])*aprob[j];
        y_train_predict[i] = np.argmax(condProb);
        if y_train[i] != y_train_predict[i]:
            trainErrorMSE = trainErrorMSE + 1;
    
    # Test Set
    y_test_predict = np.zeros(y_test.shape[0]);
    counter = np.arange(y_test.shape[0]);
    condProb = np.zeros(q);
    testErrorMSE = 0;
    for i in counter:
        for j in np.arange(q):
            condProb[j] = np.prod(prob_test[i,:,j])*aprob[j];
        y_test_predict[i] = np.argmax(condProb);
        if y_test[i] != y_test_predict[i]:
            testErrorMSE = testErrorMSE + 1;
    
    
    return [trainErrorMSE,testErrorMSE]
```

```python
# Part 8cii - Naive Bayes Classifier, Gaussian pdf

def NaiveBayesMultinomial(X_train,y_train,X_test,y_test,q): 
    # Description
    # y_train, y_test must be integers
    # X_train, X_test must be same length as corresponding y's
    # Need to filter out redundant features prior
    import numpy as np
    import pandas as pd
    
    # Step 1 : create mxs for each class to get means, std.
    #q = 6; # number of classes
    
    # Class 0 through n - training
    len_y = np.arange(y_train.shape[0]);
    for j in np.arange(q):
        expression = 'c' + str(j) + '_train = np.zeros([1,X_train.shape[1]])';
        exec(expression);
        
        expression = 'c' + str(j) + '_train = np.vstack([c' + str(j) + '_train,X_train[i,:]])'
        for i in len_y:
            if y_train[i] == j:
                exec(expression);
                
        expression = 'c' + str(j) + '_train = np.delete(c' + str(j) + '_train,0,0)'
        exec(expression);
    
    # Class 0 through n - test
    len_y = np.arange(y_test.shape[0]);
    for j in np.arange(q):
        expression = 'c' + str(j) + '_test = np.zeros([1,X_test.shape[1]])';
        exec(expression);
        
        expression = 'c' + str(j) + '_test = np.vstack([c' + str(j) + '_test,X_test[i,:]])'
        for i in len_y:
            if y_train[i] == j:
                exec(expression);
                
        expression = 'c' + str(j) + '_test = np.delete(c' + str(j) + '_test,0,0)'
        exec(expression);
  
    # Step 2 : Create a priori probabilities
    aprob = np.zeros(q);
    for j in np.arange(q):
        expression = 'aprob[j] = c' + str(j) + '_train.shape[0]/y_train.shape[0]';
        exec(expression);
    
    # Step 3 : Calculate mean and standard deviation for each conditional class - training set only!
    
    p_train = np.zeros([X_train.shape[1],2,q]); # [mean, std]
    counter = np.arange(X_train.shape[1])
    
    for j in np.arange(q): 
        for i in counter:
            expression = 'p_train[i,0,' + str(j) + '] = c' + str(j) + '_train[:,i].mean()'
            exec(expression)
            expression = 'p_train[i,1,' + str(j) + '] = c' + str(j) + '_train[:,i].std()'
            exec(expression)
        
    # Step 4 : Create probability Matrices from Training Set - assuming Conditional Gaussian
    
    condProb = np.zeros(q);
    trainErrorMSE = 0;
    testErrorMSE = 0;
    
    multi = np.zeros([X_train.shape[0],X_train.shape[1],q]);
    multi_Test = np.zeros([X_test.shape[0],X_test.shape[1],q])
    
    # Create adjusted X_train mx
    for i in np.arange(X_train.shape[1]):
        for j in np.arange(q):
            multi[:,i,j] = X_train[:,i] - p_train[i,0,j];
            multi_Test[:,i,j] = X_test[:,i] - p_train[i,0,j];
            
    y_train_predict = np.zeros(y_train.shape[0]);
        
    for i in np.arange(X_train.shape[0]):
        for j in np.arange(q):
            condProb[j] = aprob[j]*np.prod(multi[i,:,j]);
        y_train_predict[i] = np.argmax(condProb);
        if y_train[i] != y_train_predict[i]:
            trainErrorMSE = trainErrorMSE + 1;

    y_test_predict = np.zeros(y_test.shape[0]);

    for i in np.arange(X_test.shape[0]):
        for j in np.arange(q):
            condProb[j] = aprob[j]*np.prod(multi_Test[i,:,j]);
        y_test_predict[i] = np.argmax(condProb);
        if y_test[i] != y_test_predict[i]:
            testErrorMSE = testErrorMSE + 1;
    
    
    return [trainErrorMSE,testErrorMSE]
```
